```python
#MatplotLib
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# plt.show() if i use not in JUpyter
```

```python
#MUTLIPLE PLOTS

plt.subplot(1,2,1) 
#first item number of rows second item number of columns third item
#is the num of the plot
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OBJECT ORIENTED METHOD
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
```

```python
fig = plt.figure()
# create empty figure

axes1 = fig.add_axes([0.4,0.2,0.8,0.8])# left,bottom,width,hight
axes1.set_xlabel('axes1')
axes2 = fig.add_axes([0.45,0.6,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
# SUBPLOTS
```

```python
fig,axes = plt.subplots(nrows = 1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First')
axes[1].plot(y,x)
axes[1].set_title('Second')
plt.tight_layout()
# for current_ax in axes:
#     current_ax.plot(x,y)

#plt.tight_layout() #fixing overlapping
# axes.plot(x,y)
```

```python
# FIGURE SIZE AND DPI
```

```python
fig = plt.figure(figsize = (8,2))# figuresize allows us to change the size of the graph
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(2,1,figsize = (8,2))# figuresize allows us to change the size of the graph
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X Squared') # returns the label in the graph
ax.set_title('Title')
#LEGENDS
ax.plot(x,x**3,label = 'X Cubed')

ax.legend(loc=0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color ='orange',lw = 3,alpha = 1, ls = '--', marker = '*',ms = '20',
       markerfacecolor = 'yellow',markeredgewidth = 3, markeredgecolor = 'green')
# alpha is transperency lw - linewidth ls - linestyle marker - marks the points
# ms = markersize
# set the axes limits
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
